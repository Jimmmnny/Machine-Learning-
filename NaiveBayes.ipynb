```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Social_Network_Ads.csv')
```

```python
X = df.iloc[:,[2,3]].values
y = df.iloc[:,4].values
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.25,random_state=2000)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X_train,y_train)
```

```python
#Predicting the test set result 
y_pred = classifier.predict(X_test)
```

```python
#making the confusion matrix

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report

accuracy = accuracy_score(y_test,y_pred)
precision = precision_score(y_test,y_pred, average = 'weighted')
recall = recall_score(y_test, y_pred, average = 'weighted')
f1 = f1_score(y_test,y_pred, average='weighted')

print(f'Accuracy: {accuracy:.2f}')
print(f'Precision: {precision:.2f}')
print(f'Recall: {recall:.2f}')
print(f'F1 Score: {f1:.2f}')
```

```python
print("Classification Report:")
print(classification_report(y_test, y_pred))
```

Accuracy = (TP+TN)/(TP+TN+FP+FN)

定义：准确率是正确预测的比例，即模型正确预测的样本数与总样本数的比值。

适用场景：适用于类别分布均衡的情况。如果数据集中正类和负类样本的数量差不多，准确率是一个很好的评估指标。

Precision = TP/(TP+FP)

定义：精确率是指在所有被预测为正类的样本中，实际为正类的比例。

适用场景：适用于当假阳性（False Positive）代价较高的情况。例如，垃圾邮件分类中，如果将正常邮件误分类为垃圾邮件，会带来较大的负面影响。因此，需要提高精确率，减少误分类为正类的负类样本。

Recall = TP/(TP+FN)

定义：召回率是指在所有实际为正类的样本中，被正确预测为正类的比例。

适用场景：适用于当假阴性（False Negative）代价较高的情况。例如，疾病检测中，如果将患病者误分类为健康，会有严重后果。因此，需要提高召回率，确保尽可能多的正类样本被正确分类。

F1 = 2*(Precision*recall)/(Precision + Recall)

定义：F1分数是精确率和召回率的调和平均数，用来综合评价模型的性能。

适用场景：适用于需要平衡精确率和召回率的情况。当你需要一个综合考虑精确率和召回率的指标时，F1分数是一个很好的选择。
